```python
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt
import squarify

# Declaration
# Exploratory Data Analysis (EDA) - Tree Map Total Person per Job
data = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")

total_context = data.groupby('Occupation')['Person ID'].count().reset_index()

# Sort 
total_context = total_context.sort_values(by='Person ID', ascending=False)

# Extract
sizes = total_context['Person ID']
labels = total_context.apply(lambda x: f"{x['Occupation']}\n({x['Person ID']})", axis=1)

# Plot Tree Map
plt.figure(figsize=(15, 9))
squarify.plot(sizes=sizes, label=labels, alpha=0.8)
plt.title("Number of Person per Job", fontsize=14)
plt.axis('off')
plt.show()
```
