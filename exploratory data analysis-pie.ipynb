```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Most Context for Nonconditional Data
target_col = ['Gender','BMI Category','Sleep Disorder']

for dt in target_col:
    # Dataset
    top_context = df[dt].value_counts().head(7)

    # Pie Chart
    plt.figure(figsize=(8,8))
    top_context.plot.pie(autopct="%1.1f%%", startangle=90)
    plt.title(f"Most {dt}")
    plt.ylabel("")
    plt.show()
```
